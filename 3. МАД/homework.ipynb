# <center>2. Проверка данных</center>

<u>**Описание колонок**</u>

1) **id** - уникальный идентификатор для каждого пациента;

2) **age** - возраст пациента в годах;

3) **origin** - место учебы;

4) **sex** - пол: *Male/Female* (Мужской/Женский);

5) **cp** - chest pain type (тип загрудинной боли):
* *typical angina* (типичная стенокардия), 
* *atypical angina* (атипичная стенокардия), 
* *non-anginal* (неангинальная), 
* *asymptomatic* (бессимптомная);

6) **trestbps** - кровяное давление покоя, при поступлении в стационар (в мм рт.ст.);

7) **chol** - холестерин сыворотки (в мг/дл);

8) **fbs** - уровень сахара в крови натощак > 120 мг/дл:
* *True* (Да),
* *False* (Нет);

9) **restecg** - результаты электрокардиографии в состоянии покоя:
* *normal* (норма), 
* *st-t abnormality* (отклонения от нормы), 
* *lv hypertrophy* (гипертрофия ЛЖ);

10) **thalach** - достигаемая максимальная частота сердечных сокращений;

11) **exang** - стенокардия, вызванная физической нагрузкой:
* *True* (Да),
* *False* (Нет);

12) **oldpeak** - депрессия ST, вызванная нагрузкой по сравнению с отдыхом;

13) **slope** - наклон пикового сегмента ST при нагрузке;

14) **ca** - количество крупных сосудов (0–3), окрашенных при рентгеноскопии;

15) **thal**:
* *normal* (нормальный),
* *fixed defect* (исправленный дефект),
* *reversible defect* (обратимый дефект);

16) **num** - прогнозируемый атрибут:
* *0* (no heart disease / нет болезней сердца),
* *1* (mild heart disease / легкая болезнь сердца),
* *2* (moderate heart disease / умеренная болезнь сердца),
* *3* (severe heart disease / тяжелая болезнь сердца),
* *4* (critical heart disease / критическое заболевание сердца).

```python
# Избавимся от найденных выбросов
df_v2 = df_copy[df_copy['ST_depres'] >= 0]

df_v2.shape
```

```python
from scipy.stats import shapiro, levene
from scipy.stats import ttest_ind, mannwhitneyu
```

```python
# Воспользуемся функцией, которая проверяет предположения о нормальности и равной дисперсии для независимого T-теста
def check(feature):
  target = df_v2.groupby('diagnosis')
  healthy = target.get_group(0)
  patient =target.get_group(1)

  # 'isna()' выдает True, если значения нет, ставим '~' чтобы сделать реверс, получим True там, где данные в столбце есть
  healthy = healthy[~healthy[feature].isna()]
  patient = patient[~patient[feature].isna()]

  print("Предположение нормальности:")

  stat, pvalue = shapiro(healthy[feature])
  print("Здоровые:")
  print("Test statistic", stat)
  print("p-value", pvalue)

  if pvalue>0.05:
    print(f'{feature} здоровой группы нормально распределена.')
  else:
    print(f'{feature} здоровой группы распределена не нормально.')

  stat, pvalue = shapiro(patient[feature])
  print("Больные:")
  print("Test statistic", stat)
  print("p-value", pvalue)

  if pvalue>0.05:
    print(f'{feature} больной группы нормально распределена.')
  else:
    print(f'{feature} больной группы распределена не нормально.')


  stat,pvalue = levene(healthy[feature], patient[feature])
  print(f'Levene test statistic: {stat}, p-value: {pvalue}')

  if pvalue>0.05:
    print(f'Однородные.')
  else:
    print(f'Различия в {feature} между группами не однородны.')
```

```python
check('ST_depres')
```

```python
# Воспользуемся функцией, которая выполняет статистический тест (test=0 для проведения независимого T-теста, test=1 для проведения U-теста Манна-Уитни)
def stattest(feature, test):
  figure, axes = plt.subplots(1, 2, figsize=(15,5), gridspec_kw={'width_ratios':[1.5,1]})
  
  sns.histplot(data=df_v2, x=df_v2[feature], hue=df_v2['diagnosis'], palette=["lightgreen", "lightcoral"], ax=axes[0])
  axes[0].grid(alpha=0.4)
  
  sns.boxplot(data=df_v2, x=df_v2['diagnosis'], y=df_v2[feature], palette=["lightgreen", "lightcoral"], ax=axes[1])
  axes[1].set_xlabel('Группы по диагнозу')
  axes[1].grid(alpha=0.4)
  
  plt.show()

# Отображение сводной статистики между двумя группами
  target = df_v2.groupby('diagnosis')
  healthy = target.get_group(0)
  patient =target.get_group(1)
  display(target[feature].describe())

  healthy = healthy[~healthy[feature].isna()]
  patient = patient[~patient[feature].isna()]

  if test == 0:
    res = ttest_ind(healthy[feature], patient[feature])
  else:
    res = mannwhitneyu(healthy[feature], patient[feature])

  test_stat = round(res[0],4)
  pvalue = round(res[1],4)

  print('Test statistics', test_stat)
  print("Test p-value", pvalue)

# Интерпретация результатов
  if pvalue< 0.05:
    print(f"Отклонить нулевую гипотезу")
  else:
    print(f"Не удалось отвергнуть нулевую гипотезу")
```

```python
stattest('ST_depres', 1)
```

**H0**: Cтатистически значимой разницы в {feature}, у людей с сердечными заболеваниями и без них, нет.

**H1**: Cтатистически значимая разница в {feature}, у людей с сердечными заболеваниями и без них, есть.

Проверив данный признак, мы отвергаем нулевую гипотезу и можем сказать, что существует статистически значимая разница в признаке `'ST_depres'` у людей с сердечными заболеваниями и без них.

```python
# Предобработка данных
from sklearn.preprocessing import (StandardScaler, OneHotEncoder)
from sklearn.compose import ColumnTransformer

scaler = StandardScaler()
ohe = OneHotEncoder()

cat_cols = ['chest_pain', 'rest_ecg']
num_cols = ['age', 'rest_bp', 'cholesterol', 'ST_depres', 'sex', 'diabetes', 'phis_ang']
all_features = cat_cols + num_cols
target = df_v2['diagnosis'].values

x = df_v2[all_features]
y = target

preprocessor = ColumnTransformer(transformers=[
    ("scaler", StandardScaler(), num_cols),
    ("ohe", OneHotEncoder(), cat_cols)
])

preprocessor
```

# <center>8. Разделение данных на обучающую и тестовую выборки</center>

```python
import numpy as np
from sklearn.model_selection import train_test_split

SEED = 42
np.random.seed(SEED)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=SEED)

x_train.shape, x_test.shape
```

```python
# Знакомим наш трансформер (preprocessor) с тренировочными данными
preprocessor.fit(x_train)
# После знакомства, преобразуем наши тренировочные данные и сохраняем в отдельную переменную
x_transform = preprocessor.transform(x_train)
```

```python
# А тестовые данные просто преобразуем, знакомить с ними мы его не будем, т.к. нам надо, чтобы "машина" их предсказала
x_transform_test = preprocessor.transform(x_test)
```

# <center>9. Обучение</center>

**1) Логистическая регрессия**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Создаем экземпляр модели
log_reg = LogisticRegression()

# Загружаем данные
log_reg.fit(x_transform, y_train)

# Прогнозирование модели
y_pred_log_reg = log_reg.predict(x_transform_test)
print('Logistic Regression')

# Расчет accuracy
print('--'*30)
log_reg_accuracy = round(accuracy_score(y_test, y_pred_log_reg) * 100, 2)
print('Accuracy', log_reg_accuracy,'%')
```

```python
# Расчет precision and recall
from sklearn.metrics import precision_score, recall_score, confusion_matrix

# Сколько из распознанных больных действительно больные
precision = precision_score(y_test, y_pred_log_reg)
# Доля распознанных больных из всех больных пациентов
recall = recall_score(y_test, y_pred_log_reg)
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print("--"*30)

# Построение confusion matrix
confusion = confusion_matrix(y_test, y_pred_log_reg)
print(confusion)

sns.heatmap(confusion, annot=True, fmt="d")
```

**TP (True Positive)** / истинно-положительные объекты ([0][0]) =52 : пациенты представляют собой класс 0 (здоровые) и алгоритм их идентифицирует как класс 0 (здоровые)

**FP (False Positive)** / ложно-положительные объекты ([0][1]) =8 : пациенты представляют собой класс 1 (больные), алгоритм их идентифицирует как класс 0 (здоровые); ошибается

**TN (True Negative)** / истинно-отрицательные объекты ([1][1]) =69 : пациенты представляют собой класс 1 (больные) и алгоритм их идентифицирует как класс 1 (больные)

**FN (False Negative)** / ложно-отрицательные объекты ([1][0]) =19 : пациенты представляют собой класс 0 (здоровые), алгоритм их идентифицирует как класс 1 (больные); ошибается
